# 1. Scopul proiectului
Scopul principal al proiectului nostru este **dezvoltarea unei solutii pentru clasificarea tumorilor cerebrale utilizand imagini de rezonanta magnetica (MRI)**.



Tumorile cerebrale sunt considerate unele dintre cele mai agresive afectiuni, afectand atat copii, cat si adulti. Acestea reprezinta aproximativ 85 − 90% din toate tumorile primare ale sistemului nervos central (SNC). Anual, ın jur de 11.700 de persoane sunt diagnosticate cu o tumora cerebrala, iar rata de supravietuire pe cinci ani pentru pacientii cu tumori cerebrale canceroase sau SNC este de aproximativ 34% pentru barbati si 36% pentru femei. Tumorile cerebrale pot fi clasificate ın functie de natura lor: tumori benigne, tumori maligne, tumori pituitare etc. O diagnosticare corecta, ımpreuna cu un plan de tratament adecvat, este esentiala pentru a ımbunatati asteptarile de viata ale pacientilor.

Cea mai eficienta metoda de detectare a tumorilor cerebrale este imagistica prin rezonanta magnetica (MRI), care genereaza o cantitate imensa de date imagistice ce sunt examinate de catre radiologi. Cu toate acestea, examinarea manuala poate fi predispusa la erori, din cauza complexitatii implicate ın evaluarea tumorilor cerebrale si a proprietatilor acestora. Imaginile medicale, ın special cele provenite din MRI, sunt esentiale ın diagnosticarea si tratamentul tumorilor cerebrale, iar utilizarea inteligentei artificiale pentru a clasifica aceste imagini poate ımbunatati semnificativ acuratetea diagnosticarii si eficienta procesului decizional. Analiza datelor ne va permite sa identificam tiparele si variabilitatile care pot influenta clasificarea. Prin utilizarea unui set de date propus pentru clasificarea tumorilor cerebrale, ne propunem sa investigam cele mai bune practici de preprocesare a imaginilor medicale, inclusiv normalizarea, redimensionarea si augmentarea, pentru a ne asigura ca modelul nostru este robust si capabil sa generalizeze bine pe date noi.

# 2.1 Brain Tumor Classification (MRI)


Setul de date Brain Tumor Classification (MRI) contine imagini de tip MRI provenite
de la pacienti si este destinat clasificarii tumorilor cerebrale ın patru categorii distincte:
1. **Glioma Tumor**: Imagini ale tumorilor gliomatoase, care afecteaza celulele gliale din creier.
2. **Meningioma Tumor**: Imagini care ilustreaza tumorile meningiomatoase, localizate ın meningele creierului.
3. **No Tumor**: Imagini de la pacienti care nu au tumori cerebrale, utilizate ca grup de control.
4. **Pituitary Tumor**: Imagini ale tumorilor pituitare, care se dezvolta ın glanda pituitara.

Setul de date include un numar semnificativ de imagini pentru fiecare categorie, facilitand astfel antrenarea si validarea algoritmilor de clasificare bazati pe ınvatare supervizata. Aceasta diversitate de imagini ajuta la ımbunatatirea acuratetei modelului si la evaluarea performantei ın conditii variate, reflectand astfel complexitatea reala
a diagnosticarii tumorilor cerebrale. Imaginile sunt etichetate corespunzator, ceea ce permite cercetatorilor si dezvoltatorilor sa dezvolte solutii automate eficiente pentru detectarea si clasificarea tumorilor cerebrale.


Setul de date Brain Tumor Classification (MRI) este descarcat de pe platforma Kaggle si vine deja ımpart, it ın subseturi de antrenare si testare. **Din setul de date destinat antrenarii**, vom realiza o ımpartire suplimentara, alocand **80% din date pentru antrenare efectiva** si **20% pentru validare**. Aceasta ımpartire este esentiala pentru dezvoltarea unui model robust.

* Setul de antrenare (80%) va fi utilizat pentru a ınvat,a modelul sa recunoasca diferitele tipuri de tumori pe baza imaginilor MRI. Aceste date vor ajuta la ajustarea ponderilor ın timpul procesului de antrenare. Atunci cand realizam ımpartirea, vom avea grija sa selectam 80% din imaginile disponibile ın fiecare categorie,
asigurandu-ne astfel ca distributia datelor ramane echilibrata si reprezentativa.
* Setul de validare (20%) va fi folosit pentru a evalua performanta modelului ın timpul procesului de antrenare. Acesta permite monitorizarea overfitting-ului, asigurandu-se ca modelul generalizeaza bine la datele pe care nu le-a vazut anterior si ajuta la selectarea variantei optime de model (vom alege varianta pentru care
atingem cele mai bune performante pentru setul de validare).
* Setul de testare, care nu este utilizat ın procesul de antrenare, va fi rezervat pentru evaluarea finala a modelului. Aceasta va oferi o masura obiectiva a capacitatii modelului de a clasifica corect imagini noi, asemanatoare celor din practica, si de a identifica tipurile de tumori pornind de la imagini pe care nu le-a vazut anterior ın timpul antrenarii.

```python
import os
import torch
from torch.utils.data import Dataset
from PIL import Image
import numpy as np
from pathlib import Path
from torchvision import transforms
from torch.utils.data import DataLoader
from torch.utils.data import random_split, Subset
import matplotlib.pyplot as plt
import albumentations
import random
import cv2
```

```python
class BrainTumorDataset(Dataset):
  def __init__(self, root_dir):
    self.root_dir = root_dir
    self.image_paths = []
    self.labels = []

    self.class_names = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']
    for label, class_name in enumerate(self.class_names):
      class_dir = os.path.join(root_dir, class_name)
      for image_name in os.listdir(class_dir):
        image_path = os.path.join(class_dir, image_name)
        self.image_paths.append(image_path)
        self.labels.append(label)
    # Maresc numarul de label-uri pentru seturile de date
    # de training si validare ale lui no_tumor pentru 3.3
    if root_dir == "Training":
      no_tumor_indices = [i for i, label in enumerate(self.labels) if label == self.class_names.index('no_tumor')]
      self.image_paths.extend([self.image_paths[i] for i in no_tumor_indices])
      self.labels.extend([self.labels[i] for i in no_tumor_indices])

  def __len__(self):
    return len(self.image_paths)

  def __getitem__(self, idx, more_transforms = None):
    image_path, label = self.image_paths[idx], self.labels[idx]
    image = Image.open(image_path)
    if more_transforms:
      for transform in more_transforms:
        if transform == "integrity":
          if image.mode != 'RGB':
            print(f"Imaginea {idx} nu este RGB.")
          image_np = np.array(image)
          if image_np.min() < 0 or image_np.max() > 255 or image_np.max() < 30:
            print(f"Imaginea {idx} are valori ale pixelilor în afara intervalului asteptat.")
          width, height = image.size
          if width != 256 or height != 256:
            print(f"Imaginea {idx} nu are dimensiunile standard, ci {width} x {height}")
          transformation = transforms.Compose([
              transforms.Resize((256, 256)),
              transforms.ToTensor(),
              transforms.Lambda(lambda image: image.repeat(3, 1, 1) if image.shape[0] == 1 else image),
              transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
              transforms.ToPILImage()
            ])
          image = transformation(image)
        elif transform == "gaussian":
          image = np.array(image)
          image = cv2.GaussianBlur(image, (5, 5), 0)
        elif transform == "normalize":
          image = np.array(image)
          image = cv2.normalize(image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
        elif transform == "clahe":
          image = np.array(image)
          clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8, 8))
          clahe_image = clahe.apply(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))
          image = cv2.cvtColor(clahe_image, cv2.COLOR_GRAY2RGB)
        elif transform == "sobel":
          image = np.array(image)
          sobelx = cv2.Sobel(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY), cv2.CV_64F, 1, 0, ksize=5)
          sobely = cv2.Sobel(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY), cv2.CV_64F, 0, 1, ksize=5)
          sobel_image = np.sqrt(sobelx**2 + sobely**2)
          sobel_image = cv2.normalize(sobel_image, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
          image = cv2.cvtColor(sobel_image, cv2.COLOR_GRAY2RGB)
        elif transform == "equalize":
          image = np.array(image)
          equalized_image = cv2.equalizeHist(cv2.cvtColor(image, cv2.COLOR_RGB2GRAY))
          image = cv2.cvtColor(equalized_image, cv2.COLOR_GRAY2RGB)
    return image, label
```

```python
train_dataset = BrainTumorDataset(root_dir='Training')
test_dataset = BrainTumorDataset(root_dir='Testing')
```

# 3.2
Pentru a permite o evaluare robusta si generalizabila a modelului, implementati o metoda
de ımpartire a setului de date de antrenare ın doua subseturi: 80% pentru antrenare si 20% pentru validare. Aceasta metoda ar trebui sa fie flexibila, pentru a putea fi utilizata ulterior ıntr-o strategie de tip cross-validation.


```python
def split_dataset(dataset, train_ratio=0.8, validation_ratio=0.2, shuffle=False):
  train_indices = []
  validation_indices = []
  for class_label in set(dataset.labels):
    class_indices = [i for i, label in enumerate(dataset.labels) if label == class_label]
    validation_size_class = int(validation_ratio * len(class_indices))
    validation_indices_class = random.sample(class_indices, validation_size_class)
    validation_indices.extend(validation_indices_class)
    train_indices.extend([i for i in class_indices if i not in validation_indices_class])

  random.shuffle(train_indices)
  train_dataset = Subset(dataset, train_indices)
  train_dataset.rootdir = 'Training'
  train_dataset.labels = [dataset.labels[i] for i in train_indices]
  train_dataset.image_paths = [dataset.image_paths[i] for i in train_indices]
  if shuffle:
    random.shuffle(validation_indices)
  validation_dataset = Subset(dataset, validation_indices)
  validation_dataset.rootdir = 'Training'
  validation_dataset.labels = [dataset.labels[i] for i in validation_indices]
  validation_dataset.image_paths = [dataset.image_paths[i] for i in validation_indices]

  return train_dataset, validation_dataset
```

```python
train_dataset, validation_dataset = split_dataset(train_dataset, train_ratio=0.8, validation_ratio=0.2)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```
